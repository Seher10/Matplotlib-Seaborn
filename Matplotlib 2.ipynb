# Pie Chart

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Pear"]
plt.figure(figsize=(7, 7))
plt.pie(x, labels = mylabels, labeldistance=0.5, autopct="%.3f")

plt.show()
```

```python
z = np.array([16, 21, 45, 19])
mylabels = ["Apples", "Bananas", "Cherries", "Pear"]

myexplode = [0.1, 0, 0, 0]
```

```python
plt.figure(figsize=(7, 7))

plt.pie(z, labels=mylabels, explode=myexplode , labeldistance=0.75, autopct="%.1f", startangle=45);
plt.show()                                        #startangle ne yazarsan oradan başlıyor.Saat yönü tersi şeklinde başlıyor
```

```python
plt.figure(figsize=(10, 8))

color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0.5, 0.5, 0.5 ]
mylabels = ["Apples", "Bananas", "Cherries", "Pear"]

y = np.array([25, 35, 25, 15])

plt.pie(y, labels=mylabels, explode=explode_list, labeldistance=0.75, autopct="%.1f", startangle=90, shadow = True, 
        pctdistance=0.6, colors=color_list)

plt.legend(loc=(0))
plt.show()
```
